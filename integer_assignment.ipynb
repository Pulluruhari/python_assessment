```python
# Question 5: Calculate the sum of digits in number 12345
#
num = 12345
sums = 0
while num > 0:
    last_digit = num % 10
    sums = sums + last_digit
    num = num // 10
print(sums)
```

```python
# Question 6: Check if 97 is a prime number

num = 97
is_prime = True
if num <=1:
    is_prime = False
else:
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

if is_prime:
    print("prime number")
else:
    print("not a prime number")
```

```python
# Question 7: Find the factorial of 8

num = 8
fact = 1
for i in range(1, num + 1): #range(1, num + 1) â†’ range(1, 9)
    fact *= i
print(fact)
```
