```python
# Question 1: Create a set of first 10 natural numbers
natural_numbers_set = set(range(1, 11))
print(natural_numbers_set)
```

```python
# Question 2: Add element 11 to set {1, 2, 3, 4, 5}
natural_numbers_set = {1, 2, 3, 4, 5}
natural_numbers_set.add(11)
print(natural_numbers_set)
```

```python
# Question 3: Remove element 3 from set {1, 2, 3, 4, 5, 6}
natural_numbers_set = {1, 2, 3, 4, 5, 6}
natural_numbers_set.discard(3)
print(natural_numbers_set)
```

```python
# Question 4: Find the intersection of {1, 2, 3, 4, 5} and {4, 5, 6, 7, 8}
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}
intersection = set1 & set2
print(intersection)
```

```python
# Question 5: Find the difference between {1, 2, 3, 4, 5} and {4, 5, 6, 7, 8}
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}
difference = set1 - set2
print(difference)
```

```python
# Question 6: Check if 5 is in set {1, 2, 3, 4, 6, 7, 8}
natural_numbers_set = {1, 2, 3, 4, 6, 7, 8}
is_5_present = 5 in natural_numbers_set
print(is_5_present)
```

```python
# Question 7: Find the length of set {'a', 'b', 'c', 'd', 'e'}
letters_set = {'a', 'b', 'c', 'd', 'e'}
length = len(letters_set)
print(f"Length of the set: {length}")
```

```python
# Question 8: Create a set of vowels from string "hello world"
vowels = {'a', 'e', 'i', 'o', 'u'}
string = "hello world"
vowel_set = set()
for char in string:
    if char in vowels:
        vowel_set.add(char)
print(vowel_set)
```

```python
# Question 9: Remove duplicates from list [1, 2, 2, 3, 4, 4, 5, 6, 6, 7] using set
original_list = [1, 2, 2, 3, 4, 4, 5, 6, 6, 7]
unique_set = set(original_list)
print(unique_set)
```

```python
# Question 10: Check if {1, 2, 3} is a subset of {1, 2, 3, 4, 5, 6}
set_a = {1, 2, 3}
set_b = {1, 2, 3, 4, 5, 6}
is_subset = set_a.issubset(set_b)
print(is_subset)
```
